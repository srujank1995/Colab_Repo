### Assignment 6: List Methods

Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.



```python
import random 
num = [random.randint(1,20) for _ in range(25)]
asc = sorted(num)
dec = sorted(num, reverse=True)
uni = sorted(list(set(num)))

print("random list:", num)
print(f"Random list of length :{len(num)} ")
print(f"Ascending order:", asc)
print(f"Decending order:", dec)
print(f"Unique order:", uni)
```

### Assignment 8: List of Dictionaries

Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. Sort the list of dictionaries by the 'score' in descending order and print the sorted list.



```python
stud = [
     {'name': 'Alice', 'score': 88},
    {'name': 'Bob', 'score': 72},
    {'name': 'Charlie', 'score': 95},
    {'name': 'David', 'score': 65},
    {'name': 'Eve', 'score': 78}
]

sort_stud = sorted(stud,key=lambda x:x["score"], reverse=True)
print("Sorted students by score in Decending order:")
for student in sort_stud:
    print(student)

print("\n")

sort_stud = sorted(stud, key=lambda x: x['score'], reverse=False)
print("Sorted students by score in Ascending order:")
for student in sort_stud:
    print(student)
```
