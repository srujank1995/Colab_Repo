```python
## Type & Type conversion
age = 15
name= 'test'
height = 6.23

print(type(age))
print(type(name))
print(type(height))

# int ==> float
print(type(age))
print("int ==> float:",type(int(height)))

# float ==> int
print(type(height))
print("float ==> int:",type(float(age)))
```

```python
## Dynamic Typing 
var = 10 
print(var, type(var))
var = "name"
print(var, type(var))
var = 11.345 
print(var, type(var))
```

```python
## Arithmatic operator 

A = int(input("Enter the no."))
B = int(input("Enter the no."))

add = A + B
sub = A - B
div = A / B
floor = A // B
mult = A * B 
exp = A** B 
mod = A % B 

print(" Addition :", add)
print(" Substraction :", sub)
print(" Division :", div)
print(" Multiplication :", mult)
print(" Floor Div :", floor)
print(" Modulus :", mod)
print(" Exponential :", exp)
```
